```python
df["Time"] = df.Time.apply(lambda x: x[:19]).astype('datetime64')
```

```python
df["Screw Speed"] = df["Screw Speed"].apply(lambda x: np.nan if x < 0 else x)
df["Gas Flow Rate"] = df["Gas Flow Rate"].apply(lambda x: np.nan if x < 0 else x)
df["Steam Flow Rate"] = df["Steam Flow Rate"].apply(lambda x: np.nan if x < 0 else x)
df["Steam Pressure"] = df["Steam Pressure"].apply(lambda x: np.nan if x < 0.030 else x)
df["Oven-Home Temperature"] = df["Oven-Home Temperature"].apply(lambda x: np.nan if x < 24 else x)
df["Water Temperature"] = df["Water Temperature"].apply(lambda x: np.nan if x < 24 else x)
df["Oxygen_pct"] = df["Oxygen_pct"].apply(lambda x: np.nan if x < 0 else x)
df["Oven-Home Pressure"] = df["Oven-Home Pressure"].apply(lambda x: np.nan if x < -25 else x)
df["Combustion Air Pressure"] = df["Combustion Air Pressure"].apply(lambda x: np.nan if x < 0 else x)
df["Temperature before prear"] = df["Temperature before prear"].apply(lambda x: np.nan if x < 24 else x)
df["Temperature after prear"] = df["Temperature after prear"].apply(lambda x: np.nan if x < 24 else x)
df["Burner_pct"] = df.apply(lambda row: np.nan if row["Burner_pct"] < 0 else row["Burner_pct"], axis=1)
df["Borra Flow Rate"] = df["Borra Flow Rate"].apply(lambda x: np.nan if x < 0 else x)
df["Cisco Flow Rate"] = df["Cisco Flow Rate"].apply(lambda x: np.nan if x < 0 else x)
```

```python
df["Screw Speed"] = df["Screw Speed"].apply(lambda x: np.nan if x > 60 else x)
df["Gas Flow Rate"] = df["Gas Flow Rate"].apply(lambda x: np.nan if x > 1500 else x)
df["Steam Flow Rate"] = df["Steam Flow Rate"].apply(lambda x: np.nan if x > 40000 else x)
df["Steam Pressure"] = df["Steam Pressure"].apply(lambda x: np.nan if x > 25 else x)
df["Oven-Home Temperature"] = df["Oven-Home Temperature"].apply(lambda x: np.nan if x > 1000 else x)
df["Water Temperature"] = df["Water Temperature"].apply(lambda x: np.nan if x > 120 else x)
df["Oxygen_pct"] = df["Oxygen_pct"].apply(lambda x: np.nan if x > 30 else x)
df["Oven-Home Pressure"] = df["Oven-Home Pressure"].apply(lambda x: np.nan if x > 20 else x)
df["Combustion Air Pressure"] = df["Combustion Air Pressure"].apply(lambda x: np.nan if x > 150 else x)
df["Temperature before prear"] = df["Temperature before prear"].apply(lambda x: np.nan if x > 320 else x)
df["Temperature after prear"] = df["Temperature after prear"].apply(lambda x: np.nan if x > 280 else x)
df["Burner_pct"] = df.apply(lambda row: np.nan if row["Burner_pct"] > 100 else row["Burner_pct"], axis=1)
```

```python
def mva(dataframe, col_dfaltantes, col_analizar, bines=20):
    '''
    mva(missing values analysis) Tabular de acuerdo a cada columna el comportamiento
    de los valores faltantes de cierta columna.
    '''
    df_new = dataframe.copy()
    df_new['intervalos'] = pd.cut(df_new[col_analizar], bins = bines)
    df_new2 = df_new[col_dfaltantes].isnull().groupby([df_new['intervalos']]).sum().astype(int).reset_index(name='count')
    df_new2['frecuencia'] = df_new2['count']/df_new.groupby([df_new['intervalos']]).count()[col_analizar].values
    df_new2['centro_bin'] = df_new2['intervalos'].apply(lambda x: x.mid)
    return df_new2
```

```python
columnas = ["Screw Speed","Gas Flow Rate", "Steam Flow Rate", "Steam Pressure", "Oven-Home Temperature",
            "Water Temperature", "Oxygen_pct", "Oven-Home Pressure", "Combustion Air Pressure",
           "Temperature before prear", "Temperature after prear", "Burner_pct"]
```

```python
i = 0
plt.figure(figsize=(18,32))
for columna in columnas:
    ax = plt.subplot(6, 2,i+1)
    data = mva(df, columna, 'Time', bines=34)
    plt.scatter(data['centro_bin'], data['frecuencia'])
    plt.title(f'Frequency Data Atypical {columna}')
    plt.xlabel('Date')
    plt.ylabel('Frecuency')
    plt.xticks(rotation=90)
    i+=1

plt.tight_layout()
plt.savefig('frequency_plot.jpg')
plt.show()
```
